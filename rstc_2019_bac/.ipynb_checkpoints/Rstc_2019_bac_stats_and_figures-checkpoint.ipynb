# Alpha diversity

```python
def alpha_div(qza):  
  a = !unzip $qza
  digest = a[1].split('/')[0].replace('  inflating: ','')
  inf = digest + '/data/alpha-diversity.tsv'
  data = pd.read_csv(inf, sep='\t',index_col=0)
  !rm -r $digest
  return data 

alpha = pd.read_csv('metadata.tsv', sep='\t', index_col='#SampleID')
diversity = {'observed_features':'Observed ASVs',
             'shannon':'Shannon\'s entropy',
             'faith_pd':'Faith\'s PD',
             'evenness':'Pielou\'s evenness'}

for div in diversity:
  vector = 'Results/Core-metrics/%s_vector.qza' % div
  metric = div
  if div == 'shannon':
    metric = 'shannon_entropy'
  if div == 'evenness':
    metric = 'pielou_evenness'
  alpha_df = alpha_div(vector)
  alpha = pd.concat([alpha, alpha_df[metric]], axis=1)
alpha.index.name = '#SampleID'
alpha = alpha[alpha['faith_pd'].notna()]
alpha.to_csv('Results/Core-metrics/alpha.tsv',sep='\t')
```

### Day 7 and 13, Treatments and heat pumps

```python
alpha = pd.read_csv('Results/Core-metrics/alpha.tsv',sep='\t',index_col='#SampleID')
diversity = {'observed_features':'Observed ASVs',
             'shannon_entropy':'Shannon\'s entropy',
             'faith_pd':'Faith\'s PD',
             'pielou_evenness':'Pielou\'s evenness'}
for d in ['d7','d13']:
  alp_d = alpha.loc[alpha.Day == d].copy()
  for div in diversity:
    for col in ['BS_Trt','BS_Src']:
      p = stats.kruskal(*[g[div].values for n,g in alp_d.groupby(col)])[1]
      if p > 0.05:
        print('**********************************')
        print('Kruskal-Wallis p-value at day %s for %s and %s >0.05 and equals %s'%(d,col,div,p))
      if p < 0.05:
        print('**********************************')
        print('Kruskal-Wallis p-value at day %s for %s and %s <0.05 and equals %s'%(d,col,div,p))
        print('Post hoc pairwise test (Dunn’s test)')
        phoc = dunn(alp_d,val_col=div,group_col=col,p_adjust='fdr_bh')
        phoc = phoc.style.applymap(lambda x: "background-color: red" if x<0.05 else "background-color: grey")
        display(phoc)
```

# Beta diversity

```python
def beta_div(qza):  
  a = !unzip $qza
  digest = a[1].split('/')[0].replace('  inflating: ','')
  inf = digest + '/data/distance-matrix.tsv'
  data = pd.read_csv(inf, sep='\t',index_col=0)
  !rm -r $digest
  return data 

def to_matrix(df1,df2):
  df1 = df1[df1.index.isin(df2.index)]
  df1 = df1[df2.index]
  df1.sort_index(inplace=True)
  df1 = df1.reindex(sorted(df1.columns),axis=1)
  matrix = DistanceMatrix(df1,ids=df1.index)
  return matrix

def permanova_pairwise(beta,col):
  matrix = to_matrix(beta.copy(),met_d)
  p = permanova(matrix,met_d,column=col,permutations=999)[5]
  perm_res = pd.DataFrame(columns=['Group1','Group2','p'])
  if p > 0.05:
    print('Permanova p-value >0.05 and equals %s'%p)
    return p, perm_res
  if p < 0.05:
    print('Permanova p-value <0.05 and equals %s'%p)
    print('Post hoc pairwise test (Dunn’s test)')
    for pair in combinations(set(met_d[col].tolist()),2):
      metapairs = met_d.loc[(met_d[col]==pair[0])|(met_d[col]==pair[1])].copy()
      matrix2 = to_matrix(beta.copy(),metapairs)
      pp = permanova(matrix2,metapairs,column=col,permutations=999)[5]
      perm_res.loc[len(perm_res)]=[pair[0],pair[1],pp]
    perm_res['q'] = multipletests(perm_res.p,method='fdr_bh')[1]      
    return p, perm_res

meta = pd.read_csv('Results/Core-metrics/alpha.tsv', sep='\t', index_col='#SampleID')
diversity = {'unweighted_unifrac':'Unweighted UniFrac',
             'weighted_unifrac':'Weighted UniFrac',
             'jaccard':'Jaccard',
             'bray_curtis':'Bray-Curtis'}
```

### Day 7 and 13, Treatments and heat pumps

```python
days = ['d7','d13']
cols = ['BS_Trt','BS_Src']
for d in days:
  met_d = meta.loc[meta.Day == d].copy()
  for div in diversity:
    for col in cols:
      print('**********************************')
      print(d, div, col)
      beta = beta_div('Results/Core-metrics/%s_distance_matrix.qza'%div)
      p,perm_res = permanova_pairwise(beta,col)
      if p  < 0.05:
        perm_res = perm_res.loc[perm_res['Group1'].str[:-1]==perm_res['Group2'].str[:-1]]
        perm_res = perm_res.style.applymap(lambda x: "background-color: red" if x<0.05 \
                   else "background-color: grey",subset=pd.IndexSlice[:, ['p', 'q']])  
        display(perm_res)
```
