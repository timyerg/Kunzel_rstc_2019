# Alpha diversity

```python
def alpha_div(qza):  
  a = !unzip $qza
  digest = a[1].split('/')[0].replace('  inflating: ','')
  inf = digest + '/data/alpha-diversity.tsv'
  data = pd.read_csv(inf, sep='\t',index_col=0)
  !rm -r $digest
  return data 

alpha = pd.read_csv('metadata.tsv', sep='\t', index_col='#SampleID')
diversity = {'observed_features':'Observed ASVs',
             'shannon':'Shannon\'s entropy',
             'faith_pd':'Faith\'s PD',
             'evenness':'Pielou\'s evenness'}

for div in diversity:
  vector = 'Results/Core-metrics/%s_vector.qza' % div
  metric = div
  if div == 'shannon':
    metric = 'shannon_entropy'
  if div == 'evenness':
    metric = 'pielou_evenness'
  alpha_df = alpha_div(vector)
  alpha = pd.concat([alpha, alpha_df[metric]], axis=1)
alpha.index.name = '#SampleID'
alpha = alpha[alpha['faith_pd'].notna()]
alpha.to_csv('Results/Core-metrics/alpha.tsv',sep='\t')

def kw_dunn(df,col):
  p = stats.kruskal(*[g[div].values for n,g in df.groupby(col)])[1]
  phoc = pd.DataFrame(columns=['Group1','Group2','p','q'])
  if p > 0.05:
    print('Kruskal-Wallis p-value >0.05 and equals %s'%p)
  if p < 0.05:
    print('Permanova p-value <0.05 and equals %s'%p)
    print('Post hoc pairwise test (Dunn’s test)')
    dunn_p = dunn(df,val_col=div,group_col=col)
    dunn_q = dunn(df,val_col=div,group_col=col,p_adjust='fdr_bh')
    for i,pair in enumerate(combinations(set(df[col].tolist()),2)):
      phoc.loc[len(phoc)]=[pair[0],pair[1],dunn_p.loc[pair[0],pair[1]],dunn_q.loc[pair[0],pair[1]]]
  return p, phoc
    #phoc = phoc.style.applymap(lambda x: "background-color: red" if x<0.05 else "background-color: grey")
    #display(phoc)

def color_df(df):
  df = df.style.applymap(lambda x: "background-color: red" if x<0.05 \
       else "background-color: grey",subset=pd.IndexSlice[:, ['p', 'q']])
  return df
    
diversity = {'observed_features':'Observed ASVs',
             'shannon_entropy':'Shannon\'s entropy',
             'faith_pd':'Faith\'s PD',
             'pielou_evenness':'Pielou\'s evenness'}
```

### Day 7 and 13, Treatments and heat pumps

```python
alpha = pd.read_csv('Results/Core-metrics/alpha.tsv',sep='\t',index_col='#SampleID')
days = ['d7','d13']
cols = ['BS_Trt','BS_Src']
for d in days:
  df = alpha.loc[alpha.Day == d].copy()
  for div in diversity:
    for col in cols:
      print('**********************************')
      print(d, div, col)
      p,phoc = kw_dunn(df,col)
      if p < 0.05:
        phoc1 = color_df(phoc.loc[phoc['Group1'].str[:-1]==phoc['Group2'].str[:-1]].copy())
        display(phoc1)
        if col == 'BS_Trt':
          print('E vs FL within the same treatments')
          phoc2 = phoc.loc[phoc['Group1'].str[-1:]==phoc['Group2'].str[-1:]].copy()
          phoc2 = phoc2.loc[(((phoc['Group1'].str[:2]=='E_')|(phoc['Group1'].str[:2]=='FL'))&\
                                ((phoc['Group2'].str[:2]=='E_')|(phoc['Group2'].str[:2]=='FL')))|\
                                (((phoc['Group1'].str[:2]=='FR')|(phoc['Group1'].str[:2]=='FL'))&\
                                ((phoc['Group2'].str[:2]=='FR')|(phoc['Group2'].str[:2]=='FL')))].copy()
          display(color_df(phoc2))
```

```python
import numpy as np
import pandas as pd
import ecopy as ep

def beta_div(qza):  
  a = !unzip $qza
  digest = a[1].split('/')[0].replace('  inflating: ','')
  inf = digest + '/data/distance-matrix.tsv'
  data = pd.read_csv(inf, sep='\t',index_col=0)
  !rm -r $digest
  return data 

def ecopy_nmds(dist,col):
  MDS = ep.MDS(dist,naxes=2,transform='monotone')
  stress=round(MDS.stress,3)
  scores_df=pd.DataFrame(MDS.scores,index=dist.columns)
  scores_df.columns = ['NMDS1','NMDS2']
  met = pd.read_csv('metadata.tsv', sep='\t', index_col='#SampleID')
  scores_df['BS']=[met.loc[n,'BS'] for n in dist.index]
  scores_df['Day']=[met.loc[n,'Day_num'] for n in dist.index]
  scores_df['Trt']=[met.loc[n,'Treatment'] for n in dist.index]
  scores_df['HP']=[met.loc[n,'Source'] for n in dist.index]
  mdict = {'E':'^', 'FL':'o', 'FR':'X', 'RF':'d', 'RSP':'s', 'SAM':'*'}
  hue,s = 'Trt','Day'
  cdict = {'Trt1':'grey','Trt2':'blue','Trt3':'red','Trt4':'green','Trt5':'black','not_appl':'white'}
  if 'Src' in col:
    hue = 'HP'
    cdict = {'HP1':'red','HP2':'blue','not_appl':'white'}   
  elif 'Dh' in col: 
    s = 'Hour'
    scores_df['Hour']=[int(met.loc[n,'Day_hour'][-2:]) for n in dist.index]
  fig,ax=plt.subplots(figsize=(2.5,2.5),dpi=300)
  sns.scatterplot(x='NMDS1',y='NMDS2',data=scores_df,ax=ax,size=s,style='BS',\
                  markers=mdict,hue=hue,palette=cdict)
  ax.legend(bbox_to_anchor=(1,1),fontsize=6,title=None,frameon=False,markerscale=0.7)
  ax.tick_params(axis='both', length=0, pad=0)       
  ax.set_xticklabels([])
  ax.set_yticklabels([])
  ax.set_ylabel('NMDS2',fontdict={'fontsize': 6})
  ax.set_xlabel('NMDS1',fontdict={'fontsize': 6})
  ax.text(0.01,0.01,'Stress '+str(stress),size=4,transform=ax.transAxes)
  
def to_matrix(df1,df2):
  df1 = df1[df1.index.isin(df2.index)]
  df1 = df1[df2.index]
  df1.sort_index(inplace=True)
  df1 = df1.reindex(sorted(df1.columns),axis=1)
  matrix = DistanceMatrix(df1,ids=df1.index)
  return matrix, df1

def permanova_pairwise(df,col,met_d):
  matrix, df1 = to_matrix(df.copy(),met_d)
  p = permanova(matrix,met_d,column=col,permutations=999)[5]
  perm_res = pd.DataFrame(columns=['Group1','Group2','p'])
  if p > 0.05:
    print('Permanova p-value >0.05 and equals %s'%p)
  if p < 0.05:
    print('Permanova p-value <0.05 and equals %s'%p)
    print('Post hoc pairwise test (Permanova)')
    for pair in combinations(set(met_d[col].tolist()),2):
      metapairs = met_d.loc[(met_d[col]==pair[0])|(met_d[col]==pair[1])].copy()
      matrix2 = to_matrix(beta.copy(),metapairs)[0]
      pp = permanova(matrix2,metapairs,column=col,permutations=999)[5]
      perm_res.loc[len(perm_res)]=[pair[0],pair[1],pp]
    perm_res['q'] = multipletests(perm_res.p,method='fdr_bh')[1]      
  return p, perm_res, df1

def mk_biplot(distances,day,div,col):
  distances.to_csv('Results/Biplots/Filtered/%s_%s_%s_distances.tsv'%(day,div,col),sep='\t')
  dist = 'Results/Biplots/Filtered/%s_%s_%s_distances.tsv'%(day,div,col)
  matr = 'Results/Biplots/Filtered/%s_%s_%s_distances.qza'%(day,div,col)
  pcoa = 'Results/Biplots/Filtered/%s_%s_%s_pcoa.qza'     %(day,div,col)
  relt = 'Data/Relative_tables/%s_%s_%s_rel_table.qza'    %(day,div,col)
  bipl = 'Results/Biplots/Filtered/%s_%s_%s_biplot.qza'   %(day,div,col)
  bipv = 'Results/Biplots/Filtered/%s_%s_%s_biplot.qzv'   %(day,div,col)
          
  !qiime tools import \
    --input-path $dist \
    --output-path $matr \
    --type DistanceMatrix
  !qiime diversity pcoa \
    --i-distance-matrix $matr \
    --o-pcoa $pcoa
          
  distances.index.name = '#SampleID'
  distances.to_csv('Results/Biplots/Filtered/%s_%s_%s_distances.tsv'%(day,div,col),sep='\t')
          
  !qiime feature-table filter-samples \
    --i-table Data/Relative_tables/full-relative_table.qza \
    --m-metadata-file $dist \
    --o-filtered-table $relt
  !qiime diversity pcoa-biplot \
    --i-pcoa $pcoa \
    --i-features $relt \
    --o-biplot $bipl
  !qiime emperor biplot \
    --i-biplot $bipl \
    --m-sample-metadata-file metadata.tsv \
    --p-number-of-features 5 \
    --o-visualization $bipv
  
def color_df(df):
  df = df.style.applymap(lambda x: "background-color: red" if x<0.05 \
       else "background-color: grey",subset=pd.IndexSlice[:, ['p', 'q']])
  return df
          
meta = pd.read_csv('Results/Core-metrics/alpha.tsv', sep='\t', index_col='#SampleID')
diversity = {'unweighted_unifrac':'Unweighted UniFrac',
             'weighted_unifrac':'Weighted UniFrac',
             'jaccard':'Jaccard',
             'bray_curtis':'Bray-Curtis'}
!mkdir Results/Biplots/Filtered
```

### Day 7 and 13, Treatments and heat pumps

```python
days = ['d7','d13']
cols = ['BS_Trt','BS_Src']
for d in days:
  met_d = meta.loc[meta.Day == d].copy()
  for div in diversity:
    for col in cols:
      print('**********************************')
      print(d, div, col)
      beta = beta_div('Results/Core-metrics/%s_distance_matrix.qza'%div)
      p,perm_res, distances = permanova_pairwise(beta,col,met_d)
      if p  < 0.05:
        perm_res1 = perm_res.loc[perm_res['Group1'].str[:-1]==perm_res['Group2'].str[:-1]].copy()
        display(color_df(perm_res1))
        
        if col == 'BS_Trt':
          print('E vs FL within the same treatments')
          perm_res2 = perm_res.loc[perm_res['Group1'].str[-1:]==perm_res['Group2'].str[-1:]].copy()
          perm_res2 = perm_res2.loc[(((perm_res2['Group1'].str[:2]=='E_')|(perm_res2['Group1'].str[:2]=='FL'))&\
                                ((perm_res2['Group2'].str[:2]=='E_')|(perm_res2['Group2'].str[:2]=='FL')))|\
                                (((perm_res2['Group1'].str[:2]=='FR')|(perm_res2['Group1'].str[:2]=='FL'))&\
                                ((perm_res2['Group2'].str[:2]=='FR')|(perm_res2['Group2'].str[:2]=='FL')))].copy()
          display(color_df(perm_res2))
        
        if min(perm_res.q.tolist()) <= 0.05:
          mk_biplot(distances,d,div,col)
          ecopy_nmds(distances,col)
          plt.show()
```

```python
from patsy import ModelDesc
from statsmodels.formula.api import ols
import statsmodels.api as sm

def _parse_formula(formula): # borrowed from https://github.com/qiime2/q2-longitudinal
    # head off patsy errors
    if '~' not in formula:
        raise ValueError('Formula not valid: missing tilde.\n'
                         'Enter a valid formula in format "y ~ model".')
    if ';' in formula or formula.strip()[0].isdigit():
        metric = formula.split('~')[0].strip()
    else: metric = None
    # use patsy to parse formula
    model_desc = ModelDesc.from_formula(formula)
    group_columns = set()
    for t in model_desc.rhs_termlist:
        for i in t.factors: group_columns.add(i.name())
    if metric is None:      metric = model_desc.lhs_termlist[0].name()
    return metric, group_columns

df = pd.read_csv('Results/Core-metrics/alpha.tsv', sep='\t', index_col='#SampleID')
df = df.loc[(df.Day_num!=0)&(df.BS!='SAM')&(df.BS!='FR')]
sns.set(style='ticks',rc={"lines.linewidth":0.1,"axes.linewidth":0.5,'lines.markeredgewidth':0})
sns.set_context(rc={'lines.markeredgewidth':0})
palSrc={'HP1':'brown','HP2':'green'}
palTrt={'Trt1':'green','Trt2':'grey','Trt3':'brown','Trt4':'blue','Trt5':'red'}
cols = {'Treatment':palTrt,'Source':palSrc}
diversity = {'faith_pd':'Faith\'s PD','shannon_entropy':'Shannon\'s entropy',
             'observed_features':'Observed ASVs','pielou_evenness':'Pielou\'s evenness'}
bodysites = ['E','FL']
for bs in bodysites:
  data = df.loc[df.BS==bs].copy()
  fig, axes = plt.subplots(2,4,dpi=600,figsize=(4,2),gridspec_kw={'hspace': 0.35,'wspace': 0.3})
  axs,i = [[0,0],[1,0],[0,1],[1,1],[0,2],[1,2],[0,3],[1,3]],0
  for alpha in diversity:
    formula= '%s ~ Treatment+Source' % alpha
    # borrowed from https://github.com/qiime2/q2-longitudinal
    # Grab metric and covariate names from formula
    metric, group_columns = _parse_formula(formula)
    columns  = [metric] + list(group_columns)
    cats = list(set(data.columns)-set(data._get_numeric_data().columns))
    metadata = data[columns].dropna().copy()
    # borrowed from https://github.com/qiime2/q2-longitudinal
    # Grab metric and covariate names from formula
    metric, group_columns = _parse_formula(formula)
    columns  = [metric] + list(group_columns)
    cats = list(set(data.columns)-set(data._get_numeric_data().columns))
    metadata = data[columns].dropna().copy()
    # Run anova
    lm = ols(formula, metadata).fit()
    results = pd.DataFrame(sm.stats.anova_lm(lm, typ='I')).fillna('')
    # Run pairwise t-tests with multiple test correction
    pairwise_tests = pd.DataFrame()
    for group in group_columns:
        # only run on categorical columns — numeric columns raise error
        if group in cats:
            ttests = lm.t_test_pairwise(group, method='fdr_bh').result_frame
            pairwise_tests = pd.concat([pairwise_tests, pd.DataFrame(ttests)])
    if pairwise_tests.empty:
        pairwise_tests = False
    # Plot fit vs. residuals
    metadata['residual'] = lm.resid
    metadata['fitted_values'] = lm.fittedvalues
    for col in list(group_columns):
        ax, i = axes[axs[i][0]][axs[i][1]], i+1
        sns.scatterplot(x='fitted_values',y='residual',data=metadata,ax=ax,hue=col,\
                        palette=cols[col],s=4,edgecolor=None)
        ax.tick_params(axis='both', which='major', labelsize=3, length=1, pad=1, width=0.5)
        ax.set_xlabel('fitted value ',fontsize=3.5,labelpad=0.75)
        ax.set_ylabel('residual',fontsize=3.5,labelpad=0.75)
        ax.legend(borderaxespad=0,handletextpad=0,fontsize=3,frameon=False,markerscale=0.3,bbox_to_anchor=(1.3,1),loc=1)
        if i > 2: ax.set_ylabel('')
        if alpha != 'pielou_evenness': ax.legend().remove()
        if i%2 == 0: ax.text(0.5,-0.35,diversity[alpha],ha='center',fontsize=5,transform=ax.transAxes)
        if alpha == 'faith_pd': ax.text(-0.4,0.5,col,va='center',fontsize=5,\
                                      transform=ax.transAxes,rotation='vertical')
        if True in (pairwise_tests['reject-fdr_bh'].tolist()):
            trues=[str(ind)+' (%s)'%round(pairwise_tests.at[ind,'pvalue-fdr_bh'],3) for ind \
                    in pairwise_tests.index if pairwise_tests.at[ind,'reject-fdr_bh']==True]
            if trues[0].split('-')[0] in labels: ax.text(0,1.02,', '.join(trues),fontsize=3,transform=ax.transAxes)
  fig.suptitle('BodySite: %s' % bs, fontsize=5, y=0.95)
  #plt.savefig('Figures/Anova_alpha_fitted-residuals.png', bbox_inches='tight')
```
